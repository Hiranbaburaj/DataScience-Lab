```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
from scipy import stats
from sklearn.cluster import KMeans
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

# slope -> slope
# intercept -> intercept
# correlation coefficient -> r 
# p-value -> p
# standard error of the regression -> std_err
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
    return slope * x + intercept
 
# The map function applies the myfunc function to each element of the x list, 
# generating a list of predicted y-values based on the linear regression model.
mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
# Generating sample data
np.random.seed(42)
data = np.random.rand(100, 2)  # Generating 100 random data points with 2 features

# Using KMeans with 3 clusters (you can adjust the number of clusters as needed)
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)

# Getting cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plotting the data points and cluster centers
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', edgecolor='k')  # Color points based on their cluster
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centroids')  # Marking cluster centers
plt.legend()
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
