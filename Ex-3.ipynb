```python
# 1. Draw a line in a diagram from position (1, 3) to (2, 10) then to (6, 12) and finally to position # (18, 20).
#(Mark each point with a beautiful green colour and set line colour to red and line style dotted)

xpoints = np.array([1, 2, 6, 18])
ypoints = np.array([3, 10, 12, 20])

plt.plot(xpoints, ypoints,'o:r',ms='10',mec = 'g',mfc = 'g')
plt.show()

# 'o:r' o = marker style(circle) : = line format(dotted) r = line color(red)
#ms = markercolor
#mec = markeredgecolor
#mfc = markerfacecolor
```

```python
# 2. Draw a plot for the following data

# Temperature in degree Celsius  Sales
# 12                             100
# 14                             200
# 16                             250
# 18                             400
# 20                             300
# 22                             450
# 24                             500

x = np.array([12,14,16,18,20,22,24])
y = np.array([100,200,250,400,300,450,500])

plt.plot(x, y)

plt.xlabel("Temperature in degree Celsius")# label for x axis
plt.ylabel("Sales") # label for y axis

plt.show()
```

```python
# 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable 
# label in the x axis, y axis and a title.

x = [] 
y = [] 
for line in open('/home/exam1/Desktop/HIRAN_S1MCA/Datasciencelab_files/sample1.txt', 'r'): 
    lines = [i for i in line.split()] 
    x.append(lines[0]) 
    y.append(int(lines[1])) 
      
plt.title("Students Marks") #add title
plt.xlabel('Roll Number') 
plt.ylabel('Marks')
plt.yticks(y) 
plt.plot(x, y, marker = 'o', c = 'r') 
  
plt.show()
```

```python
# 4. Write a Python program to plot two or more lines on same plot with suitable legends of each line.

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1,  marker = 'o' , label='line one')
plt.plot(x2, y2, marker = '*' , ms='10',c='y', label='line two')
plt.legend() # displays the legend
plt.show()
```

```python
# 5. Write a Python program to create multiple plots.

x1 = np.array([1, 2, 6, 18])
y1 = np.array([3, 10, 12, 20])

x2 = np.array([19,15,8,2])
y2 = np.array([5,9,11,20])

plt.subplot(1, 2, 1) #the figure has 1 row, 2 columns, and this plot is the first plot. 
plt.plot(x1,y1)

plt.subplot(1, 2, 2) #the figure has 1 row, 2 columns, and this plot is the second plot. 
plt.plot(x2,y2)

plt.show()
```
