```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Decision tree using iris dataset
iris = load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state =1)

# Create Decision Tree classifier object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifier
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test,y_pred)*100} %")

# Plot the decision tree
plt.figure(figsize= (12,8))
plot_tree(clf, feature_names=iris.feature_names, class_names = iris.target_names, filled=True,rounded=True)
plt.show()
```

```python
# Decision tree using custom dataset

data = pd.read_csv('cars.csv')
df = pd.DataFrame(data)

# Convert categorical variables to numerical
df['Color'] = df['Color'].map({'Red': 1, 'Yellow': 2, 'Blue': 3})
df['Buy'] = df['Buy'].map({'Yes': 1, 'No': 0})
df['Model'] = df['Model'].apply(lambda x: 1 if x == 'Ferrari' else 0)

# Features and target variable
X = df[['Color', 'Model_Year', 'Mileage', 'Model']]
y = df['Buy']

# Create Decision Tree classifier object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifier
clf = clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(10, 8))
plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True, rounded=True)
plt.show()
```
